```python
browser.get(link)
```

```python
SCROLL_PAUSE_TIME = 5

#getting scroll height
last_height = browser.execute_script("return document.body.scrollHeight")
```

```python
for i in range(5):
    #Scroll Down to bottom
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    #Wait to load the page
    time.sleep(SCROLL_PAUSE_TIME)
    # Calculate the new scroll height nd compare last scroll height
    new_height = browser.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height  
```

```python
#get the source code
src = browser.page_source
src
```

```python
soup = BeautifulSoup(src, 'lxml')
soup
```

```python
exp_section = soup.find('section', {'id':'experience-section'})
exp_section
```

```python
li_tags = soup.find_all('div', {'class':'display-flex flex-column full-width'})
li_tags
```

```python
len(li_tags)
```

```python
a_tags = []
for i in range(len(li_tags)):
    a_tags.append(li_tags[i].find('a'))
    
```

```python
job_titles = list(map(lambda x:x.find('h3').get_text().strip(), a_tags))
job_titles
```

```python
company = list(map(lambda x:x.find_all('p')[1].get_text().strip(), a_tags))
company
```

```python
period = list(map(lambda x:x.find_all('h4')[0].find_all('span')[1].get_text().strip(), a_tags))
period
```

```python
years_exp = list(map(lambda x:x.find_all('h4')[1].find_all('span')[1].get_text().strip(), a_tags))
years_exp
```

```python
exp_roles_resps = exp_section.find_all("div", {"class": "pv-entity__extra-details t-14 t-black--light ember-view"})
```

```python
exp_roles_resps[0]
```

```python
experience = []
for exp_roles_resp in exp_roles_resps:
    for tag in exp_roles_resp.find_all('span'):
        tag.replace_with('')
    experience.append(str(exp_roles_resp).replace("<!-- --></p><!-- --></div>","").split(';">')[1].strip())
experience
```

```python
temp_df = pd.DataFrame({'profile_url':link, 'job_title':job_titles, 'company':company, 'period':period, 'years_exp':years_exp, 'experience_description':experience})
```
